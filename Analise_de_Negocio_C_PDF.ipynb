```python
!pip install -q -U google-generativeai fpdf

import google.generativeai as genai
from fpdf import FPDF
from google.colab import files

# Configuração da chave de API
GOOGLE_API_KEY = "cole sua api key"
genai.configure(api_key=GOOGLE_API_KEY)

# Solicita o nome da empresa
nome_empresa = input("Informe o nome da empresa: ")

# Configurações do modelo
generation_config = {
    "candidate_count": 1,
    "temperature": 0.5
}

safety_settings = {
    "HARASSMENT": "BLOCK_NONE",
    "Hate": "BLOCK_NONE",
    "SEXUAL": "BLOCK_NONE",
    "DANGEROUS": "BLOCK_NONE"
}

# Inicializa o modelo e inicia o chat
model = genai.GenerativeModel(model_name="gemini-1.0-pro",
                              generation_config=generation_config,
                              safety_settings=safety_settings)

chat = model.start_chat(history=[])

# Breve Descrição da Empresa
prompt_descricao = f"Descreva brevemente a empresa {nome_empresa}, incluindo seu perfil, setor atual, principais concorrentes e linhas de negócio."
response_descricao = chat.send_message(prompt_descricao)
print(f"Descrição da empresa {nome_empresa}:")
print(response_descricao.text, "\\n")

# Análise SWOT
prompt_swot = f"Realize uma análise SWOT detalhada da empresa {nome_empresa}."
response_swot = chat.send_message(prompt_swot)
print(f"Análise SWOT da empresa {nome_empresa}:")
print(response_swot.text, "\\n")

# Análise PESTEL
prompt_pestel = f"Realize uma análise PESTEL detalhada da empresa {nome_empresa}."
response_pestel = chat.send_message(prompt_pestel)
print(f"Análise PESTEL da empresa {nome_empresa}:")
print(response_pestel.text)

# Classe para criação do PDF
class PDF(FPDF):
    def header(self):
        self.set_font('Arial', 'B', 12)
        self.cell(0, 10, 'Relatório da Empresa', 0, 1, 'C')

    def chapter_title(self, title):
        self.set_font('Arial', 'B', 12)
        self.cell(0, 10, title, 0, 1, 'L')
        self.ln(2)

    def chapter_body(self, body):
        self.set_font('Arial', '', 12)
        self.multi_cell(0, 10, body)
        self.ln()

    def add_chapter(self, title, body):
        self.add_page()
        self.chapter_title(title)
        self.chapter_body(body)

# Criação e exportação do PDF
pdf = PDF()
pdf.add_chapter(f'Descrição da empresa {nome_empresa}', response_descricao.text)
pdf.add_chapter(f'Análise SWOT da empresa {nome_empresa}', response_swot.text)
pdf.add_chapter(f'Análise PESTEL da empresa {nome_empresa}', response_pestel.text)

pdf.output('Relatorio_Empresa.pdf')
files.download('Relatorio_Empresa.pdf')
```
